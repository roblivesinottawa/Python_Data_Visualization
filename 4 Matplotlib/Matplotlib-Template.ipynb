```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('data/table_i702t60.csv')
```

```python
month_number = df.loc[:, 'month'].values
interest_paid = df.loc[:, 'interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

```python
plt.style.use('seaborn-whitegrid')

plt.plot(month_number, interest_paid, marker='o', linestyle='solid')
plt.plot(month_number, principal_paid)
```

```python
# object oriented plotting
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
axes.plot(month_number, interest_paid, color='blue')
axes.plot(month_number, principal_paid, color='red')
```

```python
# how to add titles and labels

plt.figure(figsize=(10, 8))
plt.plot(month_number, interest_paid, color='blue', )
plt.plot(month_number, principal_paid, color='red')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Dollars', fontsize=14)
plt.title('Interest and Principal Paid Each Month', fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
# grids
plt.figure(figsize=(10, 8))
plt.plot(month_number, interest_paid, color='k', linewidth=2)
plt.plot(month_number, principal_paid, color='r', linewidth=2)
plt.grid(axis='y', color='0.75', linestyle='--', linewidth=1)
```

```python
# plot legends
plt.figure(figsize=(8, 6))
plt.plot(month_number, interest_paid, color='k', linewidth=2, label='Interest')
plt.plot(month_number, principal_paid, color='b', label='Principal')
plt.legend(loc='best')
```

```python
# plot legends
plt.figure(figsize=(8, 6))
plt.plot(month_number, interest_paid, color='k', linewidth=2, label='Interest')
plt.plot(month_number, principal_paid, color='b', label='Principal')
plt.legend(loc=(1.02, 0), fontsize=12)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
axes.plot(month_number, interest_paid, c='k',label='Interest')
axes.plot(month_number, principal_paid, c='b', label='Principal')
axes.legend(loc=(1.02, 0), fontsize=12)
```

```python
# saving plots to files
plt.figure(figsize=(10, 6))
plt.plot(month_number, principal_paid, color='b', linewidth=2, label='Principal')
plt.plot(month_number, interest_paid, color='k', linewidth=2, label='Interest')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(left=1, right=12)
plt.ylim(bottom=0, top=700)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Dollars', fontsize=14)
plt.title('Interest and Principal Paid Each Month', fontsize=18)
plt.legend(loc=(1.02, 0), fontsize=12)

plt.savefig('images/interest_and_principal_paid_each_month.png', dpi=300)
```

```python
#  load cancer dataset
cancer_df = pd.read_csv('data/wisconsinBreastCancer.csv')
```

```python
cancer_df.head(10)
```

```python
# how many values in diagnosis column
# b = benign m = malignant
cancer_df.diagnosis.value_counts(dropna=False)
```

```python
malignant = cancer_df.loc[cancer_df.diagnosis == 'M', 'area_mean'].values
benign= cancer_df.loc[cancer_df.diagnosis == 'B', 'area_mean'].values

plt.boxplot([malignant, benign], labels=['Malignant', 'Benign'])
```

```python
# plotting using pandas
cancer_df.boxplot(column='area_mean', by='diagnosis')
```
